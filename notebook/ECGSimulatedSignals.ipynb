## Example generating simulated signals:

- first Load NeuroKit and other useful packages

```python
import neurokit2 as nk
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Simulate

```python
ecg50 = nk.ecg_simulate(duration=10, noise=0.05, heart_rate=50)
ecg100 = nk.ecg_simulate(duration=10, noise=0.01, heart_rate=100)
```

Make a Dataframe with the results of the simulation

```python
ecg_df = pd.DataFrame({"ECG_100": ecg100, "ECG_50": ecg50})
```

Visualize

```python
nk.signal_plot(ecg_df, subplots=True)
```
